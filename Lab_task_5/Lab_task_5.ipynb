1. Import Libraries

```python
import numpy as np
import pandas as pd
```

2. Part A – Data Setup

A1. Load Dataset

```python
# Load Abalone dataset

from google.colab import files
uploaded = files.upload()
column_names = [
    "Sex", "Length", "Diameter", "Height",
    "Whole_weight", "Shucked_weight",
    "Viscera_weight", "Shell_weight",
    "Rings"
]

df = pd.read_csv("abalone.data", header=None, names=column_names)

# Number of rows
print("Number of rows:", len(df))

# Column names
print("Column names:", df.columns.tolist())

# First 5 rows
print("\nFirst 5 rows:")
print(df.head())

# ---------------- CHECKPOINT ----------------
# what is input:
# Physical numeric measurements of abalones.

# what is output:
# Rings (converted to age).

# why output is numeric:
# Age is a continuous quantity and suitable for regression.
```

A2. Convert Target

```python
# Create target variable

df["Age"] = df["Rings"] + 1.5

print(df[["Rings", "Age"]].head())
```

A3. Choose Exactly 3 Numeric Features

```python
# Selecting exactly 3 features

features = ["Length", "Diameter", "Shell_weight"]

X = df[features].values
y = df["Age"].values.reshape(-1, 1)

# ---------------- JUSTIFICATION ----------------
# Feature 1: Length → captures overall size growth.
# Feature 2: Diameter → reflects maturity and shell expansion.
# Feature 3: Shell_weight → indicates accumulated biological growth.
```

A4. Train-Test Split (80/20, No Sklearn)

```python
N = len(X)
split_index = int(0.8 * N)

X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

A5. Normalize Inputs (Using Training Statistics Only)

```python
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std

# ---------------- CHECKPOINT ----------------
# why normalization is needed for learning:
# Gradient descent depends on scale.
# If features are on different scales, updates become unstable
# and learning becomes slow or divergent.
```

3. Part B – Define Linear Model (Forward Pass)

```python
def forward(X, w, b):
    """
    Computes y_hat = Xw + b
    """
    y_hat = X @ w + b

    return y_hat

# ---------------- CHECKPOINT ----------------
# parameters are: w (weights) and b (bias)
# number of parameters:
# 3 weights + 1 bias = 4 parameters
```

4. Part C – Loss Functions

MSE

```python
def mse(y, y_hat):
    return np.mean((y - y_hat) ** 2)

# ---------------- CHECKPOINT ----------------
# why square:
# Removes sign and penalizes large errors heavily.

# what mistakes are expensive:
# Large prediction errors are penalized more than small ones.
```

MAE

```python
def mae(y, y_hat):
    return np.mean(np.abs(y - y_hat))
```

5. Part D – Learning Rule (Gradients)

D1. Error Definition


e=y_hat−y

D2. Gradient Meaning


```python
# ---------------- CHECKPOINT ----------------
# what gradient means in words:
# Gradient is the slope of the loss with respect to a parameter.
# It tells how loss changes if we slightly change that parameter.

# why subtracting gradient reduces loss:
# Because gradient points toward direction of increase.
# Subtracting it moves us toward direction of decrease.
```

D3. Gradient Functions

```python
def grad_w(X, y, y_hat):
    N = len(y)
    dW = (2/N) * (X.T @ (y_hat - y))
    return dW

def grad_b(y, y_hat):
    N = len(y)
    db = (2/N) * np.sum(y_hat - y)
    return db

# ---------------- CHECKPOINT ----------------
# meaning of large gradient:
# Loss is very sensitive to that parameter.

# effect of too-large learning rate:
# Model may overshoot minimum and diverge.
```

6. Part E – Training Loop

```python
np.random.seed(42)

d = X_train.shape[1]

# Initialize parameters
w = np.random.randn(d, 1) * 0.01
b = 0.0

learning_rate = 0.01
epochs = 500

print("Initial expectation:")
print("Loss should decrease gradually over epochs.")

for epoch in range(epochs):

    # 1. Forward pass
    y_hat = forward(X_train, w, b)

    # 2. Compute loss
    loss = mse(y_train, y_hat)

    # 3. Compute gradients
    dW = grad_w(X_train, y_train, y_hat)
    db = grad_b(y_train, y_hat)

    # 4. Update parameters
    w = w - learning_rate * dW
    b = b - learning_rate * db

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("Revised expectation after training:")
print("Loss decreases steadily but does not eliminate systematic bias.")
```

7. Part F – Evaluation

```python
# Test predictions
y_test_hat = forward(X_test, w, b)

# Test MSE
test_mse = mse(y_test, y_test_hat)
print("Test MSE:", test_mse)

# Test MAE
test_mae = mae(y_test, y_test_hat)
print("Test MAE:", test_mae)
```

Print 5 Sample Predictions

```python
print("\nFive Sample Predictions:")
for i in range(5):
    true_val = y_test[i][0]
    pred_val = y_test_hat[i][0]
    abs_error = abs(true_val - pred_val)

    print(f"True Age: {true_val:.2f}, Predicted Age: {pred_val:.2f}, Absolute Error: {abs_error:.2f}")
```

8. Reflective Comments

```python
# ---------------- REFLECTION ----------------

# systematic errors:
# Older abalones tend to be underestimated.
# Younger abalones tend to be overestimated.

# observed bias:
# Predictions are pulled toward the mean age.
# The linear model creates a flat surface and cannot bend
# to capture different growth behaviors at different ages.

# key insight:
# Gradient descent successfully reduces error,
# but it cannot change the model structure.
# The limitation is representational, not optimization-based.
```
